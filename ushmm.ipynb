```python
import json
import glob
import pandas as pd
from leet_topic import leet_topic
```

```python
files = glob.glob("ushmm/*.json")
len(files)
```

```python
segments = []
for file in files:
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)
    for item in data["segments"]:
        try:
            segments.append({"text": item["question"], "file": file})
            segments.append({"text": item["answer"], "file": file})
        except:
            KeyError
len(segments)
```

```python
df = pd.DataFrame(segments, columns=["text", "file"])
df
```

```python
new_df, topic_data = leet_topic.LeetTopic(df,
                                          document_field="text",
                                          html_filename="ushmm.html",
                                          extra_fields=["file"],
                                          max_distance=.5
                                        )
```
