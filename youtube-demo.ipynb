```python
import json
import glob
import pandas as pd
from leet_topic import leet_topic
```

```python
files = glob.glob("ushmm/*.json")
len(files)
```

```python
documents = []
for file in files:
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)
    for segment in data["segments"]:
        try:
            documents.append({"text": segment["question"], "file": file})
            documents.append({"text": segment["answer"], "file": file})
        except:
            KeyError
len(documents)
```

```python
documents[:2]
```

```python
df = pd.DataFrame(documents)
df
```

```python
new_df, topic_data = leet_topic.LeetTopic(df, document_field="text", html_filename="ushmm2.html", max_distance=.5)
```
